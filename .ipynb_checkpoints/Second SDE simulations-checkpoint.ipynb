#Rearrang ing data from 3/1/2022 t o 30/12/2022

```r
n1 = round(0.8 * n, 0)
n2 = n1 + 1
n3 = n2:n
train = price[1:n1]
test = price[n2:n]
train.date = date[1:n1]
test.date = date[n2:n]
n1
```

 ##Fitting the USD/KES Train Data to the GBM Model
 

```r
data = setData(train, delta = 1/252, t0 = 0)
model = setModel(drift = "mu*x", diffusion = "sigma*x", state.var = "x", time.var = "t", solve.var = "x", xinit = train[1])
D = setYuima(model = model, data = data)
```

```r
initialise = list(mu = 0.05, sigma = 2)
lowbound = list(mu = -Inf, sigma = 0)
upbound = list(mu = Inf, sigma = Inf)
mle = qmle(D, start = initialise, lower = lowbound, upper = upbound)
summary(mle)
```

#E s t im a te d Parame ters

```r
sigma=coef (mle) [ 1 ]
mu=coef(mle) [ 2 ]
sigma 
mu
```

Simulating in-sample Paths of the fitted model


```r
J = 1000
V = matrix(0, nrow = n1, ncol = J)

for (j in 1:J) {
  grid = setSampling(delta = 1/252, n = n1-1)
  Y = simulate(model, true.param = list(mu = mu, sigma = sigma), sampling = grid)
  sim.Data = Y@data@original.data
  V[, j] = sim.Data[, ]
}

```

Train Set Estimated Mean Equation and Confidence Intervals
#G e t t i n g t h e mean e q u a t i o n from t h e 1000 f i t t e d p a t h s
Actual = Black Solid Line
Mean Path = Blue Solid Line
95% C.I = Red Dotted Line

```r
train.est = rep(0, n1)
train.uci = rep(0, n1)
train.lci = rep(0, n1)

for (i in 1:n1) {
  train.est[i] = mean(V[i, ])
  train.lci[i] = quantile(V[i, ], 0.025)
  train.uci[i] = quantile(V[i, ], 0.975)
}

par(mfrow = c(1, 1))
plot(train.est ~ train.date, type = "l", main = "Train Dataset Fit", xlab = "2022", ylab = "Lines", col = "blue")
lines(train.uci ~ train.date, col = "red", lty = 2)
lines(train.lci ~ train.date, col = "red", lty = 2)
```

Testing the Goodness of Fit of the Model
H0 : Mean simulated sample path and the actual Train data come from same distribution
H1 : Mean simulated sample path and the actual Train data come from different distributions
The Kolmogorov Smirnov test statistic is given by:
D = supx|F(x)−G(x)| (3.1)
where:
F(x) is the empirical distribtuion of the simulated mean sample path.
G(x) is the empirical distribtuion of the actual train data.


```r
ks.test(train.est, train)
```

Forecasting the Future USD/KES Prices
#Tes t D a t a se t
#The USD/KES p r i c e on 26/5/2023 was 1 3 8 . 2 5
#We need t o f o r e c a s t t h e v a l u e s f o r t h e t e s t p e r i o d

```r
model1 <- setModel(drift = expression(mu * x),
                   diffusion = expression(sigma * x),
                   state.var = "x",
                   time.var = "t",
                   solve.var = "x",
                   xinit = price[n1])

```

Simulating out-of-sample Paths of the fitted model
#S im ul a t i n g 1000 t e s t d a t a sample p a t h s from t h e f i t t e d model
#Removing t h e i n i t i a l v a l u e wh ich i s t h e l a s t v a l u e o f t h e t r a i n d a t a s e t
M=M[ =1 ,]

```r
J = 1000
m = length(test)
M = matrix(0, nrow = m + 1, ncol = J)

for (j in 1:J) {
  grid = setSampling(delta = 1/252, n = m)
  X = simulate(model1, true.param = list(mu = mu, sigma = sigma), sampling = grid)
  test.Data = X@data@original.data
  M[, j] = test.Data[, ]
}
```

Train Set Estimated Mean Equation and Confidence Intervals
#G e t t i n g t h e mean e q u a t i o n and 95% C . I from t h e 1000 f i t t e d p a t h s
#G e t t i n g t h e mean e q u a t i o n from t h e 1000 f i t t e d p a t h s

```r
test.est = rep(0, m)
test.uci = rep(0, m)
test.lci = rep(0, m)

for (i in 1:m) {
  test.est[i] = mean(M[i, ])
  test.lci[i] = quantile(M[i, ], 0.025)
  test.uci[i] = quantile(M[i, ], 0.975)
}

plot(price ~ date, type = "l", main = "Forecasted USD/KES Price Process", xlab = "2022", ylab = "Price (Ksh)")

lines(test.est ~ test.date, col = "blue")
lines(test.uci ~ test.date, col = "red", lty = 2)
lines(test.lci ~ test.date, col = "red", lty = 2)
```
