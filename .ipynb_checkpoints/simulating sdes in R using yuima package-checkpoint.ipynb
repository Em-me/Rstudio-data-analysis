# Simulating SDE'S 

```r
library(yuima)
model<-setModel(drift="0.1*x",diffusion="0.2*x",state.var="x",time.var="t",solve.var="x",xinit=100)
x<-simulate(model)
plot(x)
```

```r
x@data@original.data
```

```r
library(yuima)
theta<-0.1
model<-setModel(drift="((-theta*y)/(1+y^2)^0.5)",diffusion="1",state.var="y",time.var="t",solve.var="y",xinit=0.5)
grid<-setSampling(Terminal=100,delta=0.5)
y<-simulate(model)
plot(y)
```

```r
y@data@original.data
```

```r
library(yuima)
model<-setModel(drift="((-theta*y)/(1+y^2)^0.5)",diffusion="1",state.var="y",time.var="t",solve.var="y",xinit=0.5)
grid<-setSampling(Terminal=100,n=200)
y<-simulate(model,sampling=grid,true.param=list(theta=0.5))
plot(y)
```

```r
y@data@original.data
```
